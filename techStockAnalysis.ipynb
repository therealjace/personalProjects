```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
```

```python
tickers = ['MU', 'NVDA', 'AMD', 'AAPL', 'GOOG', 'NFLX', 'AMZN', 'FB']
tech_stocks = pd.DataFrame()
for t in tickers:
    tech_stocks[t] = wb.DataReader(t, data_source='yahoo', start='2013-1-1')['Adj Close']
```

```python
tech_stocks.head()
```

```python
(tech_stocks / tech_stocks.iloc[0] * 100).plot(figsize = (15,6))
plt.show()
```

```python
(tech_stocks / tech_stocks.iloc[0] * 100).tail()
```

```python
simple_returns = (tech_stocks / tech_stocks.shift(1)) - 1
simple_returns.tail()
```

```python
average_returns = simple_returns.mean() * 252
average_returns
```

```python
weights = np.array([.125,.125,.125,.125,.125,.125,.125,.125])
weights
```

```python
tech_portfolio = str(round(np.dot(weights, average_returns), 4) * 100)
```

```python
print ('The average annual return of this equally weighted portfolio was ' , tech_portfolio , '%')
```

```python
print ('The stock with the highest average return was', average_returns.idxmax())
```
